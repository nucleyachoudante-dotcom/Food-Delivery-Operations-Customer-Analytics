```python
import numpy as np 
import pandas as pd
```

```python
df = pd.read_csv('food_delivery_analytics_cleaned.csv')
```

```python
#revenue by city 
df.groupby('city_tier')['final_amount_paid'].sum()
```

```python
#averge delivery time
df.groupby('city_tier')['delivery_time_minutes'].mean()
```

```python
#customer rating
df.groupby('city_tier')['customer_rating'].mean()
```

```python
#promo_code impact
df.groupby('promo_code_used')['final_amount_paid'].mean()
```

```python
#delay vs rating

df.groupby('delayed_delivery_flag')['customer_rating'].mean()
```
